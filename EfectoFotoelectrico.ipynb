## Antecedentes e historia entorno al efecto fotoeléctrico

El año 1905 fue un año extraordinario para la Física y en especial para Albert Einstein quien con sólo 25 años publicó 4 estudios que cambiaron no sólo a la Física sino también al mundo como se conocía. Uno de los estudios más conocido es el de la "Relatividad Especial" pero fue gracias a la investigación que realizó del efecto Fotoeléctrico, que 16 años después recibiría el Premio Nobel de Física.

Ya en 1887, Heinrich Hertz observó que una placa al ser iluminada generaba corriente eléctrica pero hasta el momento la luz era sólo considerada una onda. En 1900 Planck habló de "cuantos" de energía pero hasta ese momento no era posible explicar porque en el experimento del efecto fotoeléctrico el aumentar la intensidad de la luz no implicaba electrones con mayor energía.

Einstein aplicó las ideas de Planck y los experimentos del efecto fotoeléctrico aplicados por Lenard. En su artículo llamado [“Sobre un punto de vista heurístico relativo a la producción y transformación de la luz”] (http://www.relatividad.org/ef_fotoel_einst_heuristico.pdf "Efecto Fotoelectrico") explica lo siguiente:

>Hasta donde yo veo, no existe contradiccion alguna entre las propiedades del efecto fotoeléctrico observados por Lenard y nuestra interpretación. Si cada cuanto de energía de la luz incidente cede su energ ıa independientemente de todos los restantes, entonces la distribución de velocidades de los electrones, esto es, la calidad de la radiación catódica generada será independiente de la intensidad de la luz excitante; por otro lado el número de electrones que dejan el sólido será proporcional a la intensidad de la luz excitante en condiciones por lo demás identicas

Es decir, Einstein demostró que la luz se puede comportar como una onda, pero también como un conjunto de partículas que llamó "cuantos de luz" (fotones). Aumentar la intesidad de la luz no es aumentar la energía de los cuantos de luz, sólo se aumenta la cantidad de cuantos de luz. En cambio, si queremos que los electrones tengan más energía se debe aumentar la frecuencia de la fuente de luz.

Este estudio no fue bien aceptado en principio, pero fue comprobado experimentalmente por Millikan quien al recibir el premio Nobel de Física dijo:

>Después de diez años de probar, cambiar, aprender y a veces equivocarse, todos los esfuerzos, habiendo estado dirigidos desde el principio a la medición experimental precisa de las energías de emisión de los fotoelectrones... este trabajo resultó, contradiciendo mis propias esperanzas, en la primera prueba experimental directa en 1914 de la validez exacta, dentro de muy estrechos límites de errores experimentales, de la ecuación de Einstein, y de la primera determinación fotoeléctrica directa de la constante h de Planck. 



## Explicación sobre el efecto fotoeléctrico
El efecto fotoeléctrico es el proceso por el cual se liberan electrones de un material por la acción de la radiación. Cada material metálico tiene una frecuencia mínima de radiación electromagnetica necesaria para que empiecen a liberarse los electrones, si no se alcanza esta frecuencia mínima por más que se incremente la intensidad de la luz no se liberarán electrones, por otro lado,  el aumentar la intensidad de la luz liberará más electrones.

## Conceptos importantes  

### Energía de arranque $\phi$ : 
Es también llamada de extracción o función de trabajo (W) y es la energía mínima necasaria para que un electrón escape del metal

### Energía cinetica en el efecto fotoeléctrico $E_k$


La energía cinética es igual a la energía del fotón menos la energía de arranque:

$$E_k = E - \phi$$


El electrón absorve la energía (E) del fotón trás el impacto de un rayo de luz, si la Energía (E) es menor que la enería de arranque ($\phi$) no se extraeran electrones. En cambio, si es mayor, el electrón escapará de la atracción del núcleo. La energía sobrante se convierte en energía cinética del elctrón.

### Energía potencial en contra

Al aplicar una diferencia de potencia V, se genera una energía potencial en contra que hace que los electrones no lleguen a la otra placa y es ahí, cuando la energía potencial ($e\cdot V$) y la Energía cinética ($E_k$) se igualan:

$$
e\cdot V = E - \phi
$$

Podemos comprobar que $e\cdot V$ ($e$ es la carga del electón) tiene unidades  de energía de la siguiente manera:

$$e = \mathrm{C} $$
$\mathrm{C = A \cdot s} $

Por lo tanto:

$$

e = \mathrm{A \cdot s}

$$

Por otro lado, el Voltio equivale a una medición de potencia ($\mathrm{W}$) sobre el amperaje ($\mathrm{A}$) del circuito: 
$$\mathrm{W = \frac{J}{s}}$$

$V = \mathrm{\frac{J}{A \cdot s}} $

$$\require{cancel} e \cdot V = (\mathrm{\cancel{A \cdot s}})(\mathrm{\frac{J}{\cancel{A \cdot s}}}) = \mathrm{J}$$

Joules es una unidad que mide la Energía

Einstein postuló que el efecto fotoeléctrico podía asimilarse a un uno-contra-uno de partículas de luz (llamadas fotones) con cada electrón. La energía de estos fotones sería $E= \hbar f$ , donde $f$ es la frecuencia de la luz (y por ende, de cada fotón) y $\hbar$ es la así llamada constante de Planck. De esta manera, podemos reescribir la ecuación anterior como:
$$eV=\hbar f - \phi$$

Conociendo el potencial de frenado y la frecuencia de la luz que impacta contra el metal podemos ajustar todos los datos a una recta donde su ordenada en el origen tendrá que ver con la función de trabajo y la pendiente estará en función justamente la constante de Planck


## Ejercicio de calentamiento

Un ajuste a una recta (o regresión lineal) lo que persigue es hallar la ecuación de la recta que mejor pasa por una serie de puntos. Recordamos que una recta se puede definir, como mínimo, con dos puntos $x_1, y_1$ e $x_2, y_2$ , de manera que cualquier otro punto de dicha recta $x , y$  cumple que:

$\frac{x-x_1}{x_2-x_1} = \frac{y-y_1}{y_2-y_1}$  y esta es equivalente a $m= \cdot x + b$

Siendo 

$m= \frac{y_1 - y_2}{x_1-x_2}$    y  $b=\frac{x_1 \cdot y_2 - x_2 \cdot y_1}{x_1-x_2}$

**Demuestra la equivalencia de la relación matemática que:**

$\frac{x-x_1}{x_2-x_1} = \frac{y-y_1}{y_2-y_1}$ es equivalente a $y=m\cdot x+b$

Para resolver el ejercicio propuesto comenzamos multiplicando los dos lados de la igualdad por $y_2-y_1$:

$$(y_2-y_1) \cdot \bigg(\frac{x-x_1}{x_2-x_1}\bigg) = \bigg(\frac{y-y_1}{y_2-y_1}\bigg) \cdot (y_2-y_1)$$

Ahora simplificamos los terminos equivalentes:
$$(y_2-y_1) \cdot \bigg(\frac{x-x_1}{x_2-x_1}\bigg) = \bigg(\frac{y-y_1}{\cancel{y_2-y_1}}\bigg)\cdot \cancel{(y_2-y_1)}\quad\Rightarrow\quad\frac{(y_2-y_1)\cdot(x-x_1)}{x_2-x_1} = y-y_1$$

Sumamos ahora $y_1$ en los dos lados para obtener:

$$\frac{(y_2-y_1)\cdot(x-x_1)}{x_2-x_1} + y_1 = y$$

A continuación multiplicamos los terminos del numerador de la fracción y multiplicamos $y_1$ por $\frac{x_2-x_1}{x_2-x_1}$, ya que la fracción vale uno y, por lo tanto, no cambia el valor de la igualdad:

$$\frac{xy_2 - xy_1 - x_1y_2 + x_1y_1}{x_2-x_1} + y_1\bigg(\frac{x_2-x_1}{x_2-x_1}\bigg) = y$$

Multiplicando $y_1$ por la fracción conseguimos juntar los dos sumandos en una sola fracción:

$$\frac{xy_2 - xy_1 - x_1y_2 + \cancel{x_1y_1} + x_2y_1\cancel{-x_1y_1}}{x_2-x_1} = y$$

Ahora sacamos factor común "$x$":

$$\frac{x(y_2 - y_1) - x_1y_2 + x_2y_1}{x_2-x_1} = y$$

Dividiendo la fracción en dos partes conseguimos ya casi la equivalencia pedida:

$$\bigg(\frac{y_2 - y_1}{x_2-x_1}\bigg)x+\frac{- x_1y_2 + x_2y_1}{x_2-x_1} = y$$

Finalmente multiplicamos todo por \frac{-1}{-1} y obtenemos la equivalencia requerida:

$$\bigg(\frac{-1}{-1}\bigg)\Bigg(\bigg(\frac{y_2 - y_1}{x_2-x_1}\bigg)x+\frac{- x_1y_2 + x_2y_1}{x_2-x_1}\Bigg) = y \bigg(\frac{\cancel{-1}}{\cancel{-1}}\bigg) \quad \Rightarrow \quad y = \bigg(\frac{y_1-y_2}{x_1-x_2}\bigg)x+\frac{x_1\cdot y_2 - x_2\cdot y_1}{x_1-x_2}$$

Donde: $$m=\frac{y_1-y_2}{x_1-x_2}\qquad \qquad b=\frac{x_1\cdot y_2 - x_2\cdot y_1}{x_1-x_2}$$

## Cálculo de la constante $\hslash$
### Ecuación de $\hslash$
Para calcular ahora $\hslash$, la despejaremos de la ecuación en el paso anterior:
$$
\hslash = \frac{m\cdot e}{2\pi \cdot c}
$$
### Ahora debemos saber sus unidades:

Para conseguirlo debemos saber que unidades posee la pendiente:
$$
m = \frac{2\pi \hslash c}{e} \qquad \Rightarrow \quad  m = \mathrm{\frac{J\cdot s\cdot \frac{m}{s}}{C}}
$$
$$
\mathrm{J = Kg\cdot \frac{m}{s^2}\cdot m = Kg\cdot \frac{m^2}{s^2}}
$$
$$
\mathrm{C = A\cdot s}
$$
$$
\Rightarrow \quad  m = \mathrm{\frac{Kg\cdot\frac{m}{s^2}\cdot m\cdot s\cdot\frac{m}{s}}{A\cdot s} = \frac{Kg\cdot m^3\cdot \cancel{s}}{A\cdot s^\cancelto{3}4} = \frac{Kg\cdot \frac{m^2}{s^2}\cdot m}{A\cdot s} = \frac{J\cdot m}{C}}
$$
Conociendo ahora las unidades de la pendiennte podemos calcular las unidades de la constante de Planck:
$$
\hslash = \frac{m\cdot e}{2\pi \cdot c} \qquad \Rightarrow \qquad \hslash = \mathrm{\frac{\frac{J\cdot \cancel{m}}{\cancel{C}}\cdot \cancel{C}}{\frac{\cancel{m}}{s}} = J\cdot s}
$$
Sabiendo que:
$$
c = 3\times10^{8}\cdot \mathrm{\frac{m}{s}}\qquad\qquad e = -1.6\times10^{-19}\cdot\mathrm{C}
$$
Podemos calcular la constante:

```javascript
var carga_el = -1.6*1e-19;
var vel_luz = 3*1e8;
constante_de_planck = (promedio*carga_el)/(2*Math.PI*vel_luz);
console.log('La constante de Planck ħ calculada es: ' + constante_de_planck +'·Js');
```

Y calculando la constante de Planck h, la cual es:
$$
h = \hslash \cdot 2\pi
$$
nos sale:

```javascript
constante_de_planck_h = constante_de_planck * 2 * Math.PI;
console.log('Y la constante de Planck h calculada es: ' + constante_de_planck_h +'·Js');
```

## Energía de arranque ($\phi$):

Como mencionado previamente la pendiente representa la constante de Planck ($\hslash$) de forma que:
$$
b = \frac{\phi}{e}
$$
Podemos observar que los valores de la ordenadas en el origen ($b$) de ambos metales no coinciden. Eso es debido a que cada metal posee una frecuencia umbral diferente.

### La frecuencia umbral:
Es la minima frecuencia que debe de tener la luz que incide sobre el metal para conseguir que el electrón se libere del núcleo. Cada metal tiene su propia frecuancia lumbral, es decir, en algunos será más facil o más difícil arrancar un electrón de su núcleo. Mientras que con una frecuencia concreta se liberan electrones en el sodio, en el cinc puede que no haga efecto ya que tiene una frecuencia umbral mayor.


## Cálculo de la energía de arranque ($\phi$):

### Ecuación de $\phi$:
Para calcular $\phi$, la despejaremos de la ecuación en el paso anterior:
$$
\phi = b\cdot e
$$
### Debemos saber sus unidades:
Para conseguirlo debemos saber que unidades posee la ordenada en el origen:
$$
b = \frac{\phi}{e} \qquad \Rightarrow \qquad b = \mathrm{\frac{J}{C}}
$$
onociendo ahora las unidades de la pendiennte podemos calcular las unidades:
$$
\phi = b\cdot e \qquad \Rightarrow \qquad \phi = \mathrm{\frac{J}{\cancel{C}}\cdot \cancel{C} = J}
$$
Sabiendo que:
$$
e = -1.6\times10^{-19}\cdot\mathrm{C}
$$
Podemos calcular la energía de arranque para el sodio y el cinc:


```javascript
energia_arranque_sodio = b_sodio * carga_el; // la carga del electrón ya fue declarada previemente
energia_arranque_cinc = b_cinc * carga_el;
console.log('La energía de arranque del sodio vale ' + energia_arranque_sodio + '·J');
console.log('La energía de arranque del cinc vale ' + energia_arranque_cinc + '·J');
```

La energías de arranque nos salen negativas porque accionan al contra de la intensión nuestra, de arrancar el electrón.

#### ¿Qué hace CheepJ tras el telón y en qué estándares Web se basa?
CheepJ convierte automaticamente con anticipación y ejecuta Java al estandar HTML5, de esta manera el usuario no necesita tener una instalación local y puede correr la aplicación o la librería en cualquier navegador. CheepJ cuenta con funciones basadas en el navegador coo sistemas de archivos virtualizados, redes y audio.

Trás el telón lo que hace CheepJ es convertir cualquier aplicación Java en HTML5/WebAssembly/JavaScript.

#### ¿Qué son los CDN?
CDN (Content Delivery Network) o redes de distribución son servidores que se encuentran en diferentes ubicaciones que generan puntos de presencia fuera del servidor de origen. Los beneficios de los CDN son:

- Los puntos de presencia ayudan a bajar los costos en el ancho de banda, facilitando el acceso a los datos  con la ayuda de herramientas de optimización como la memoría caché.

- La seguridad se incrementa cuando se usa CDN ya que tienen altos estandares de autenticación y cifrado además de que desvían el tráfico del servidos orginal a servidores proxy y esto evita ataques en la seguridad del servidor.

- Mejora la distribución del contenido por medio de diferentes sistemas de entrega
